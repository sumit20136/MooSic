```python
import keras
import tensorflow as tf
import cv2
import numpy as np
```

```python
model=tf.keras.models.load_model('/content/drive/MyDrive/ML_Project/model.h5')
```

```python
#/content/drive/MyDrive/ML_Project/train
Datadirectory='/content/drive/MyDrive/ML_Project/test'
# path=os.path.join(Datadirectory,"")
# print(path)
Classes=['happy','neutral','sad']
```

```python
import os
testing_data=[]
def create_testing_data():
  for Category in Classes:
    path=os.path.join(Datadirectory,Category)
    class_num=Classes.index(Category)
    for img in os.listdir(path):
      img_arr=cv2.imread(os.path.join(path,img))
      new_array=cv2.resize(img_arr,(224,224))
      testing_data.append([new_array,class_num])
create_testing_data()
```

```python
X=[]
Y=[]
for features,label in testing_data:
  X.append(features)
  Y.append(label)
X=np.array(X).reshape(-1,224,224,3)
```

```python
Y_pred=(model.predict(X))
```

```python
import pandas as pd
pd.DataFrame(Y_pred[4])
```

```python
max = Y_pred.max(axis=1, keepdims=True)
max_index = Y_pred.argmax(axis=1)

print('Max:', max)
print('Max Index:', max_index)
```
